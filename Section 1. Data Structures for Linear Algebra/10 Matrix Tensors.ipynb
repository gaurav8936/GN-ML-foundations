### Matrices (Rank 2 Tensors) in NumPy

```python
import numpy as np
```

```python
X = np.array([[25, 2], [5, 26], [3, 7]])

# notice the capital X
```

```python
X
```

```python
X.shape
```

```python
X.size

# total number of elements in the array
```

```python
# Select left column of matrix X (zero-indexed)

X[:,0]

# 0 is used to represent the first column
# Python index starts at 0
```

```python
# Select middle row of matrix X: \

X[1,:]
```

```python
# Another slicing-by-index example: 

# first to second row and first to second column

X[0:2, 0:2]
```

### Matrices in TensorFlow

```python
import tensorflow as tf
```

```python
X_tf = tf.Variable([[25, 2], [5, 26], [3, 7]])
```

```python
X_tf
```

```python
tf.rank(X_tf)
```

```python
# rank0 is scalar
# rank1 is vector
# rank2 is matrix
```

<font color = red>  rank0 is scalar,  rank1 is vector,  rank2 is matrix

```python
tf.shape(X_tf)
```

```python
# Slicing
```

```python
X_tf[1,:]
```
