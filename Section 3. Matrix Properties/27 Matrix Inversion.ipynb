```python
import numpy as np
```

```python
X = np.array([[4,2], [-5, -3]])
```

```python
X
```

```python
Xinv = np.linalg.inv(X)
```

```python
# the .inv is for matrix inversion
```

```python
Xinv
```

```python
y = np.array([4, -7])
```

```python
y
```

***

```python
w = np.dot(Xinv, y)
```

```python
w
```
