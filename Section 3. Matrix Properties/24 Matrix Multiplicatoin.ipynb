## PyTorch

```python
import torch
```

```python
A_pt = torch.tensor([[3,4], [5,6], [7,8]])
```

```python
b_pt = torch.tensor([1, 2])
```

```python
torch.matmul(A_pt, b_pt)
```

```python
# this automatically infers dims in order to perform dor product, matvec, ormatric mult
```

## TensorFlow

```python
import tensorflow as tf
```

```python
A_tf = tf.Variable([[1,2], [3,4], [5,6]])
```

```python
b_tf = tf.Variable([1,2])
```

```python
tf.linalg.matvec(A_tf, b_tf)
```

***

```python
B_tf = tf.convert_to_tensor(B, dtype=tf.int32)

# this is like the from numpy method we used in Pytorch
```

```python
tf.matmul(A_tf, B_tf)
```
