### Orthogonal Matrices

These are the solutions to Exercises 3 and 4 on **orthogonal matrices** from the slides.

For Exercise 3, to demonstrate the matrix $I_3$ has mutually orthogonal columns, we show that the dot product of any pair of columns is zero: 

```python
I = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
I
```

```python
column_1 = I[:,0]
column_1
```

```python
column_2 = I[:,1]
column_2
```

```python
column_3 = I[:,2]
column_3
```

```python
np.dot(column_1, column_2)
```

```python
np.dot(column_1, column_3)
```

```python
np.dot(column_2, column_3)
```

***

We can use the `np.linalg.norm()` method from earlier in the notebook to demonstrate that each column of $I_3$ has unit norm: 

```python
np.linalg.norm(column_1)
```

```python
np.linalg.norm(column_2)
```

```python
np.linalg.norm(column_3)
```

<font color = red> Since the matrix $I_3$ has mutually orthogonal columns and each column has unit norm, the column vectors of $I_3$ are *orthonormal*. Since $I_3^T = I_3$, this means that the *rows* of $I_3$ must also be orthonormal. 

<font color = red> Since the columns and rows of $I_3$ are orthonormal, $I_3$ is an *orthogonal matrix*.

```python
# (1) each of the pair of columns are orthogonal to each other (product is zero)
```

```python
# (2) each of the columns has unit norm (L2 norm - squared sum of all elements) = 1
```
