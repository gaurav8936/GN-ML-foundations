```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each. 

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]] 
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
        
    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

$$ \lim_{x \to \infty} \frac{25}{x} $$

```python
def inf_fxn(my_x1):
    my_y1 = 25/my_x1
    return my_y1
```

```python
x1 = np.linspace(-10, 10, 10000) # start, finish, n points
```

```python
y1 = inf_fxn(x1)
```

```python
inf_fxn(1e3)

# 1e3 (scientific notation for 1000)
```

```python
inf_fxn(1e6)
```

```python
# this is now really close to 0
```

***

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
_ = ax.plot(x1, y1)
```

```python
# quirky behavior as matplotlib connects the points

# we cannot evaluate at 0

# the two parts of the function are not connected
```

```python
# in order to resovle this

# we split this into two arrays
```

```python
left_x = x[x<0]
```

```python
right_x = x[x>0]
```

```python
left_y = inf_fxn(left_x)
right_y = inf_fxn(right_x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
ax.plot(left_x, left_y, c='C0')
_ = ax.plot(right_x, right_y, c='C0')
```
